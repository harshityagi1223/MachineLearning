### **Cleaning Amazon Fashion Data**

```python
import pandas as pd
import numpy as np
path='/content/gdrive/MyDrive/CA683- Main Assignment/AMAZON-FASHION-json.xls'
amazon_fashion = pd.read_excel(path)
```

```python
amazon_fashion.head()
```

```python
amazon_fashion.isna().sum()
```

```python
amazon_fashion[amazon_fashion.columns[0]].count()
```

```python
amazon_fashion.drop(['_ - style - Color:'], axis=1,inplace=True)
```

```python
amazon_fashion.drop(['_ - style - Size:','_ - style - Style:','_ - style - Size Name:','_ - style - Metal Type:','_ - style - Length:','_ - style - Team Name:','_ - style - Style Name:','_ - style - Format:','_ - style - Package Quantity:','_ - style - Gem Type:','_ - style - Item Package Quantity:','_ - style - Shape:','_ - style - Scent Name:','_ - style - Material:','_ - style - Item Display Length:','_ - image - image'], axis=1,inplace=True)
```

```python
amazon_fashion.isna().sum()
```

```python
#we delete rows all values as NA 
```

```python
amazon_fashion.dropna(how = 'all', axis=0,inplace=True)
```

```python
amazon_fashion.isna().sum()
```

```python
amazon_fashion.to_excel('AmazonFasionUpdated.xlsx', index=False)
```
